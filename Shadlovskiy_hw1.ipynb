# Завдання 2

```python
import pandas as pd

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

tables = pd.read_html(url)

target_table = tables[20]
print(target_table)
```

```python
target_table.head()
```

```python
df = pd.DataFrame(target_table)
shape = df.shape
print("Кількість рядків і стовпців:", shape)
```

```python
df.replace(np.nan,'—', inplace=True)
df
```

```python
column_types = df.dtypes

print("Типи стовпців:")
print(column_types)
```

```python
non_numeric_columns = df.select_dtypes(exclude=[np.number]).columns
column_to_remove = 'регіон'
on_numeric_columns = non_numeric_columns.drop(column_to_remove, errors='ignore')
print(on_numeric_columns)
df['1950']=pd.to_numeric(df['1950'],errors='coerce')
df['1960']=pd.to_numeric(df['1960'],errors='coerce')
df['1970']=pd.to_numeric(df['1970'],errors='coerce')
df['2014']=pd.to_numeric(df['2014'],errors='coerce')
column_types = df.dtypes
print(column_types)
```

```python
missing_counts = df.isnull().sum()

print("Кількість пропущених значень у кожному стовпці:")
print(missing_counts)
```

```python
df = df.drop(df.index[-1])
df
```

```python
df = df.fillna(df.mean())
df
```

```python
average_birth_rate_2014 = df['2014'].mean()
regions_with_higher_birth_rate = df[df['2014'] > average_birth_rate_2014]
region_list = regions_with_higher_birth_rate['регіон'].tolist()
print(region_list)
```

```python
max_2012=df['2012'].max()
max_2008=df['2008'].max()
reg_2012 = df[df['2012']==max_2012]["регіон"].tolist()
reg_2008 = df[df['2008']==max_2008]["регіон"].tolist()
print("Максимальна народжуваність в 2012:",reg_2012[0])
print("Максимальна народжуваність в 2008:",reg_2008[0])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))  
plt.bar(df['регіон'], df['2014'], color='skyblue') 
plt.xlabel('Регіон')  
plt.ylabel('Народжуваність в 2014 році') 
plt.title('Народжуваність за регіонами в 2014 році') 
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()  
plt.show() 
```
