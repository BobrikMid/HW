```python
import altair as alt
from vega_datasets import data
wheat_data = data.wheat()
```

```python
wheat_data.head()
```

```python
source = data.wheat()

bar_chart = alt.Chart(source).mark_bar().encode(
    x='year:N',
    y='wheat:Q',
    color=alt.condition(
        alt.datum.year == 1810,
        alt.value('orange'),
        alt.value('steelblue')
    ),
    tooltip=['year:N', 'wheat:Q']
).properties(width=600)


bar_chart
```

```python
bar_chart = alt.Chart(source).mark_bar().encode(
    x='year:N',
    y='wheat:Q',
    color=alt.value('steelblue'), 
    tooltip=['year:N', 'wheat:Q']
)

average_rule = alt.Chart(source).mark_rule(color='red').encode(
    y='mean(wheat):Q'
)

combined_chart = alt.layer(bar_chart, average_rule).properties(width=600)

combined_chart = combined_chart.encode(
    color=alt.value('steelblue'),
    tooltip=['year:N', 'wheat:Q']
).properties(
    title='Стовпчаста діаграма виробництва пшениці',
    description='Середнє значення позначено червоним правилом'
)

combined_chart.save('wheat_production_chart.html')
```

```python
stocks_data = data.stocks()

interactive_chart = alt.Chart(stocks_data).mark_line(point=True).encode(
    x='date:T',
    y='price:Q',
    color='symbol:N'
).properties(width=800)

interactive_chart = interactive_chart.encode(tooltip=['date:T', 'price:Q'])
interactive_chart.save('stocks_interactive_chart.html')
```
