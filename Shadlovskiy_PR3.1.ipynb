##                                                              Візуалізація Matplotlib 

```python
import pandas as pd
df = pd.read_csv("C:/Users/faika/OneDrive/Desktop/spotify-2023.csv", encoding="latin1")
```

```python
df.head
```

```python
import matplotlib.pyplot as plt
top_songs = df.sort_values(by='streams', ascending=False).head(10)
top_songs = top_songs.iloc[1:]
plt.figure(figsize=(15, 10))
plt.barh(top_songs['track_name'], top_songs['streams'], color='skyblue')
plt.xlabel('Кількість прослуховувань')
plt.title('10 найпопулярніших пісень на Spotify у 2023 році')
plt.gca().invert_yaxis() 
plt.show()
```

```python
year_counts = top_songs['released_year'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(year_counts, labels=year_counts.index, autopct='%1.1f%%')
plt.title('Розподіл року випуску серед найпопулярніших пісень у 2023 році')
plt.show()
```

### Теж немає данних :)

```python
top_artists = df.sort_values(by='streams', ascending=False).head(10)
top_artists = top_artists.iloc[1:]
plt.figure(figsize=(10, 6))
plt.barh(top_artists['artist(s)_name'], top_artists['streams'], color='skyblue')
plt.xlabel('Кількість слухачів')
plt.title('10 найпопулярніших артистів на Spotify у 2023 році')
plt.gca().invert_yaxis()  
plt.show()
```

```python
tracks_to_compare = ["I Love You So", "Where Are You Now", "Seek & Destroy"]

filtered_data = df[df['track_name'].isin(tracks_to_compare)]


plt.figure(figsize=(10, 6))
plt.barh(filtered_data['track_name'], filtered_data['streams'], color='skyblue')
plt.xlabel('Кількість прослуховань')
plt.title('Порівняння популярності треків у 2023 році')
plt.gca().invert_yaxis()  
plt.show()
```
